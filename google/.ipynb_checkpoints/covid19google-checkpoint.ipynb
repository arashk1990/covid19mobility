```python
data.dropna(axis=1, how='all',inplace = True)
```

```python
data.drop(data[data['sub_region_1'].isna()==True].index,inplace = True)
```

```python
# sub-region 1 corresponds to 152 UK counties/cities as listed in:
#https://en.wikipedia.org/wiki/Districts_of_England
print(len(data['sub_region_1'].unique()))
data['sub_region_1'].unique()
```

```python
# sub-region 2 corresponds to 268 UK districts as (some of them) listed in:
#https://en.wikipedia.org/wiki/Districts_of_England
print(len(data['sub_region_2'].unique()))
data['sub_region_2'].unique()
```

```python
#Second-level subdivisions
# https://en.wikipedia.org/wiki/ISO_3166-2:GB
print(len(data['iso_3166_2_code'].unique()))
data['iso_3166_2_code'].unique()
```

```python
# limit to greater london
gl = data[data['sub_region_1'].str.contains('London')]

gl.drop(gl[gl['sub_region_2'].isna()==True].index,inplace = True)

gl.head()
```
